Executive visualizations with seaborn

```python
sns.set_theme(style="whitegrid", palette="deep")
plt.rcParams['figure.figsize'] = (10,6)
```

Executive Barchart

```python
#which room type dominates, and how do they differ in booking beahaviour
sns.countplot(
    x='room_type',
    data=df,
    hue='is_bookable'
)

plt.title('Room Type Distribution by Booking Availability', fontsize=14, weight='bold')
plt.xlabel('Room Type')
plt.ylabel('Number of Listings')

plt.legend(title='Instant Bookable')
plt.show()
```

Executive piechart

```python
#what share does each room type contribute?
counts = df['room_type'].value_counts()

colors = sns.color_palette("pastel")

plt.pie(
    counts,
    labels=counts.index,
    autopct='%1.1f%%',
    colors=colors,
    startangle=90
)

plt.title('Market Share of Room Types', fontsize=14, weight='bold')

plt.show()
```

Executive line graph

```python
#How does pricing scale with capacity?
sns.lineplot(
    x='accommodates',
    y='log_price',
    data=df,
    estimator='mean',
    marker='o'
)

plt.title('Average Price by Capacity', fontsize=14, weight='bold')
plt.xlabel('Number of People Accommodated')
plt.ylabel('Average Log Price')

plt.show()
```

Executive histogram

```python
#Are prices concentrated or sprayed out?
sns.histplot(
    df['log_price'],
    kde=True,
    bins=10
)

plt.title('Distribution of Listing Prices', fontsize=14, weight='bold')
plt.xlabel('Log Price')
plt.ylabel('Frequency')

plt.show()
```

```python

sns.scatterplot(
    x='accommodates',
    y='log_price',
    data=df,
    hue='room_type',
    size='bedrooms',
    sizes=(50, 200),
    alpha=0.7
)

plt.title('Price vs Capacity by Room Type', fontsize=14, weight='bold')
plt.xlabel('Accommodates')
plt.ylabel('Log Price')

plt.legend(title='Room Type / Bedrooms')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="darkgrid", context="talk")

g = sns.catplot(
    x="room_type",
    hue="is_bookable",  # Changed from 'instant_bookable' to 'is_bookable'
    col="cancellation_policy",
    data=df,
    kind="count",
    height=5,
    aspect=1
)

g.fig.suptitle("Room Type Distribution by Booking & Cancellation Policy", y=1.05)
g.tight_layout()

plt.show()
```

```python
#pie chart
counts = df['room_type'].value_counts()
colors = sns.color_palette("coolwarm")

plt.pie(
    counts,
    labels=counts.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    wedgeprops={'width':0.4}
)

plt.title("Market Share of Room Types", fontsize=16, weight='bold')
plt.show()
```

line graph

```python
sns.lineplot(
    x='accommodates',
    y='log_price',
    data=df,
    estimator='mean',
    ci='sd',
    marker='o'
)

plt.title("Price Trend by Capacity (with Variation)", fontsize=16, weight='bold')
plt.xlabel("Capacity")
plt.ylabel("Average Log Price")

plt.show()
```

```python
#Histogram
sns.histplot(
    data=df,
    x='log_price',
    hue='room_type',
    kde=True,
    bins=12,
    multiple='stack'
)

plt.title("Price Distribution by Room Type", fontsize=16, weight='bold')
plt.xlabel("Log Price")

plt.show()
```
